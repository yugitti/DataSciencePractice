```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
custom_style = {'axes.labelcolor': 'white',
                'xtick.color': 'white',
                'ytick.color': 'white'}
sns.set_style("darkgrid", rc=custom_style)
plt.style.use('dark_background')

%precision 3
%matplotlib inline
```

```python
sampling_x = np.random.uniform(-1, 1, 10000)
sampling_y = np.random.uniform(-1, 1, 10000)
```

```python
plt.scatter(sampling_x, sampling_y, marker='.', color='yellowgreen')
```

```python
#
dist = np.sqrt(sampling_x**2 + sampling_y**2)
sampling_x_outer = sampling_x[dist > 1]
sampling_y_outer = sampling_y[dist > 1]
sampling_x_inner = sampling_x[dist <= 1]
sampling_y_inner = sampling_y[dist <= 1]
```

```python
plt.figure(figsize=(5,5))
plt.scatter(sampling_x_outer, sampling_y_outer, marker='.', color='yellowgreen')
plt.scatter(sampling_x_inner, sampling_y_inner, marker='.', color='tomato')
```

```python
circle_area_sample = len(sampling_x_inner) 
rectangle_area_sample = len(sampling_x)
```

```python
circle_area = circle_area_sample / rectangle_area_sample
```
