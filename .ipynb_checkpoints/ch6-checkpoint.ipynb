```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data1 = {
    'id': ['100', '101', '102', '103', '104', '106', '108', '110', '111', '113'],
    'city': ['Tokyo', 'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo', 'Tokyo', 'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo'],
    'birth_year': [1990, 1989, 1992, 1997, 1982, 1991, 1988, 1990, 1995, 1981],
    'name':['Hiroshi', 'Akiko', 'Yuki', 'Satoru', 'Steeve', 'Mituru', 'Aoi', 'Tarou','Suguru', 'Mitsuo']
}
data2 = {
    'id': ['100', '101', '102', '105', '107'],
    'math': [50, 43, 33, 76, 98],
    'english': [90, 30, 20, 50, 30],
    'index_num': [0, 1, 2, 3, 4]
}
```

```python
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

#### merge

```python
## inner merge
pd.merge(df1, df2, on='id', how='inner')
```

```python
# outer merge
pd.merge(df1, df2, on='id', how='outer')
```

```python
# left merge
pd.merge(df1, df2, on='id', how='left')
```

```python
# right merge
pd.merge(df1, df2, on='id', how='right')
```

#### map

```python
# 地域名の列を追加
city_map = {'Tokyo': 'Kanto', 'Hokkaido': 'Hokkaido', 'Osaka': 'Kansai', 'Kyoto': 'Kansai'}
```

```python
df1['region'] = df1['city'].map(city_map)
```

```python
df1
```

```python
# 年齢を追加
df1['age'] = df1['birth_year'].map(lambda x: 2019 - x)
```

```python
df1
```

#### ビン分割

```python
# 誕生年を５年区切りに分割
bins = [1980, 1985, 1990, 1995, 2000]
birth_year_cut_data = pd.cut(df1['birth_year'], bins=bins)
print(birth_year_cut_data)
```

```python
birth_year_cut_data.value_counts()
```

```python
#  分割数を指定
pd.cut(df1['birth_year'], 2)
```

```python
#  分位点で分割 -> ほぼ同じサイズになるように分割
pd.qcut(df1['birth_year'], 2).value_counts()
```

#### group_by

```python
df1.groupby(['region','city'], as_index=True).mean()
```

```python
df1.groupby(['region','city'], as_index=False).mean()
```

```python
#groupbyの返り値はイテレーターに対応したメソッド
for gp, subdf in df1.groupby('region'):
    print("===============================")
    print("region name: {}".format(gp))
    print(subdf)
```

#### 時系列データ

```python
import pandas_datareader.data as pdr
```

```python
start_date = "2001/1/2"
end_date = "2016/12/30"
```

```python
fx_data = pdr.DataReader("DEXJPUS", "fred", start_date, end_date)
```

```python
fx_data.head()
```

```python
plt.plot(fx_data["DEXJPUS"])
```

```python
#特定の年月
fx_data["2016-4"].head()
```

```python
# 月初のデータをピック
test = fx_data.resample('M')
fx_data.resample('M').first().head()
```

```python
# 月末のデータをピック
fx_data.resample('M').last().head()
```

```python
# 月の平均をピック
fx_data.resample('M').mean().head()
```

```python
# 欠損穴埋め
fx_data.resample('D').last().head()
```

```python
# 直前のデータで穴埋め
fx_data.resample('D').ffill()
```

```python
# 直後のデータで穴埋め >> 連続してNaNだと穴埋めされない
fx_data.resample('D').bfill()
```

```python
# 変化率を計算
(fx_data / fx_data.shift(1)).head()
```
