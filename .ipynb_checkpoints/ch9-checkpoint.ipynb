```python
#  エルボー法によるクラスター数の推定（重心からの距離の総和）
max_cluster = 20
dist = []
for i in range(1, max_cluster+1):
    km = KMeans(init="random", n_clusters=i, random_state=0)
    km.fit(X)
    dist.append(km.inertia_)
```

```python
# シルエット係数
X, _ = make_blobs(random_state=10)
plt.scatter(X[:,0], X[:,1])
```

```python
km = KMeans(init="random", n_clusters=3, random_state=0)
c = km.fit_predict(X)
```

```python
from sklearn.metrics import silhouette_samples, pairwise_distances
```

```python
s = silhouette_samples(X, c)
xx = np.c_[X, c,s]
labels = np.unique(xx[:,2])
y_min, y_max = 0, 0
s_avg = xx[:,3].mean()
for l in labels:
    a = xx[xx[:,2] == l]
    a = a[np.argsort(a[:,3])]
    y_max = len(a) + y_min
    plt.barh(range(y_min, y_max), a[:,3], height=1, edgecolor='none')
    y_min = y_max
plt.axvline(s_avg, linestyle='--')
```

```python
D = pairwise_distances(X)
D0 = D[:, c==0]
D1 = D[:, c==1]
D2 = D[:, c==2]
Ds = [D0, D1, D2]
clusters = np.unique(c)
ss = np.zeros(len(X))
for cluster in clusters:
    s = np.array([])
    s_temp = []
    targets = np.delete(clusters, cluster)
    d_current = np.sum(Ds[cluster], axis=1) / (Ds[cluster].shape[1]-1)
    for t in targets:
        t_dist = np.sum(Ds[t], axis=1) / Ds[t].shape[1]
        s_temp = (t_dist - d_current) / np.max(np.c_[t_dist, d_current], axis=1)
        if len(s) == 0:
            s = s_temp
        else:
            s = np.vstack((s, s_temp))
    s = np.min(s, axis=0)
    ss[c==cluster] = s[c==cluster]
```

```python
def pair_distance(X):
    H = np.tile(np.diag(np.dot(X, X.T)), (np.size(X.T, 1), 1))
    G = np.dot(X, X.T)
    return np.sqrt(H - 2*G + H.T)
```
