```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_context('notebook', font_scale=0.9)
%precision 3
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
X, _ = make_blobs(random_state=10)
plt.scatter(X[:,0], X[:,1])
```

```python
km = KMeans(init="random", n_clusters=3)
km.fit(X)
y_pred = km.predict(X)
```

```python
df = pd.DataFrame(X)
df['cluster'] = y_pred
df.columns = ["X0", 'X1', "cluster"]
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="X0", y="X1", s=200, data=df, hue="cluster", legend="full")
plt.legend()
```
